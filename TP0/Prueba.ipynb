```python
import random
import math
import matplotlib.pyplot as plt

random.seed(42)
tamanios = [10, 50, 100, 500]

plt.figure(figsize=(9, 6))
for n in tamanios:
    normas = []
    for _ in range(n):
        # vector aleatorio en R^2 con componentes en [0,1)
        x = random.random()
        y = random.random()
        normas.append(math.sqrt(x*x + y*y))

    # índice de cada punto (1..n)
    xs = list(range(1, n+1))
    # puntos sin unir:
    plt.scatter(xs, normas, s=18, label=f"{n} puntos")

plt.title("Norma de vectores aleatorios en R²")
plt.xlabel("Índice del punto")
plt.ylabel("‖v‖")
plt.ylim(0, math.sqrt(2))
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import random, math
import matplotlib.pyplot as plt
random.seed(48)

tamanios = [10, 50, 100, 500]
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = [ax for row in axes for ax in row]

for ax, n in zip(axes, tamanios):
    normas = []
    for _ in range(n):
        x = random.random(); y = random.random()
        normas.append(math.sqrt(x*x + y*y))
    ax.scatter(range(1, n+1), normas, s=18)
    ax.set_title(f"{n} puntos")
    ax.set_ylim(0, math.sqrt(2))
    ax.set_xlabel("Índice"); ax.set_ylabel("‖v‖")
    ax.grid(True, alpha=0.3)

plt.suptitle("Norma de vectores aleatorios en R²", y=1.02)
plt.tight_layout()
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

tamanios = [10, 50, 100, 500]
normas = []

for n in tamanios:
    # Generar UN vector en R^n
    v = [random.random() for _ in range(n)]
    # Calcular su norma euclidiana
    norma = math.sqrt(sum(x*x for x in v))
    normas.append(norma)

# Graficamos: tamaño del vector vs su norma
plt.figure(figsize=(8,5))
plt.plot(tamanios, normas, marker='o', linestyle='-', color='b')
plt.title("Norma de un vector aleatorio en R^n")
plt.xlabel("Dimensión del vector (n)")
plt.ylabel("‖v‖")
plt.grid(True)
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

tamanios = [10, 50, 100, 500]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = [ax for row in axes for ax in row]

for ax, n in zip(axes, tamanios):
    # Generar un vector en R^n
    v = [random.random() for _ in range(n)]
    norma = math.sqrt(sum(x*x for x in v))

    # Graficar la norma (un único valor)
    ax.bar([f"R^{n}"], [norma], color="skyblue", edgecolor="black")
    ax.set_ylim(0, math.sqrt(n))  # rango teórico máximo
    ax.set_title(f"Norma de un vector en R^{n}")
    ax.set_ylabel("‖v‖")

plt.suptitle("Norma de vectores aleatorios en distintas dimensiones", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

tamanios = [10, 50, 100, 500]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = [ax for row in axes for ax in row]

for ax, n in zip(axes, tamanios):
    # Generar un vector en R^n
    v = [random.random() for _ in range(n)]
    norma = math.sqrt(sum(x*x for x in v))

    # Graficar solo el valor de la norma
    ax.plot([n], [norma], marker="o", markersize=10, color="blue")
    ax.set_ylim(0, math.sqrt(n))  # rango teórico máximo
    ax.set_title(f"Norma en R^{n}")
    ax.set_xlabel("Dimensión n")
    ax.set_ylabel("‖v‖")

    # Mostrar en consola (primeros 10 componentes)
    print(f"\nVector en R^{n}:")
    if n > 10:
        print(v[:10], "...")  # solo primeros 10
    else:
        print(v)
    print(f"Norma en R^{n}: {norma:.4f}")

plt.suptitle("Norma de un vector aleatorio en distintas dimensiones", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```
