# Ejercicio 1

Dada la ecuación de la recta, se solicita emplear la librería Matplotlib (o alguna de su elección) para graficar a través de una función los puntos de la misma.

```python
print('Configurar parámetros de la recta.')
cant_elementos = 20
a = 2
b = 10
recta = [a*x+b for x in range(cant_elementos)]

print(f"Valores de la recta: {recta}")
```

Tarea: Usar la librería de Matplotlib para graficar la recta.

```python
x_values_recta = [i for i in range(cant_elementos)]
plt.plot(x_values_recta,recta,'ro-',label=f'recta simple a={a} - b={b}')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title(f"Recta en el plano: pendiente {a} y ordenada al origen {b}.")
```

# Ejercicio 2

Graficar la norma del vector para una corrida con 10, 50, 100, y 500 puntos aleatorios. De ser necesario, buscar lo que se denomina "norma de un vector".

```python
import random
import math

tamanios = [10, 50, 100, 500]


plt.figure(figsize=(10, 6))

for n in tamanios:
    normas = []
    for i in range(n):
        # Generamos un vector aleatorio en R^2
        v = []
        for j in range(2):
            v.append(random.random())
        
        # Calculo la norma del vector generado.
        suma_cuadrados = 0
        
        for x in v:
            suma_cuadrados += x * x
        norma = math.sqrt(suma_cuadrados)
        
        normas.append(norma)
    
    # Graficamos la lista de normas para esta corrida
    plt.plot(range(1, n + 1), normas, marker='o', linestyle='-', label=f"{n} puntos")

    plt.title("Norma de vectores aleatorios en R^2")
    plt.xlabel("Valores de X")
    plt.ylabel("Valores de Y")
    plt.grid(True)
    plt.legend()
    plt.show()
```
